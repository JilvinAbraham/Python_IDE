```python
from tkinter import *
import os

class PythonIDE:
    
    def __init__(self):
    
        win = Tk()
        win.title("Python_IDE")
        win.geometry("1200x650")
        
        editor_label = Label(text = "\t\t\t\t\tWRITE YOUR CODE HERE")
        editor_label.place(x=40,y=25)
        
        #code editor part
        self.code_editor=Text(win,relief=GROOVE,state=NORMAL,bd=3,width=80,height=38)
        self.code_editor.place(x=40,y=50)
        
        #save button
        save_btn = Button(text = "Save Code",relief=GROOVE,command = self.read_code)
        save_btn.place(x = 40,y=677)
        
        #run button
        run_btn = Button(text = "Run Code",relief=GROOVE,command = self.run)
        run_btn.place(x = 122,y=677)
        
        #input 1 part        
        inp_1_lbl = Label(text = "Input 1")
        inp_1_lbl.place(x=720,y=50)
        self.input1=Text(win,relief=GROOVE,state=NORMAL,bd=3,width=30,height=15)
        self.input1.place(x=720,y=80)
        inp1_save_btn = Button(text = "Save 1",relief=GROOVE,command = self.read_input1)
        inp1_save_btn.place(x = 820,y=340)
        
        
        #input 2 part
        inp_2_lbl = Label(text = "Input 2")
        inp_2_lbl.place(x=720,y=380)
        self.input2=Text(win,relief=GROOVE,state=NORMAL,bd=3,width=30,height=15)
        self.input2.place(x=720,y=417)
        inp2_save_btn = Button(text = "Save 2",relief=GROOVE,command = self.read_input2)
        inp2_save_btn.place(x = 820,y=675)
        
        
        #output 1 and output 2
        outp_1_lbl = Label(text = "Output 1")
        outp_1_lbl.place(x=1020,y=50)
        self.output1=Text(win,relief=GROOVE,state=NORMAL,bd=3,width=30,height=15)
        self.output1.place(x=1020,y=80)
        
        outp_2_lbl = Label(text = "Output 1")
        outp_2_lbl.place(x=1020,y=380)
        self.output2=Text(win,relief=GROOVE,state=NORMAL,bd=3,width=30,height=15)
        self.output2.place(x=1020,y=417)
        

        win.mainloop()
        
    #function to read code from text widget and save it in a .py file
    def read_code(self):
        code = self.code_editor.get("1.0",END)
        f1 = open("runner1.py","w+")
        f1.write("""import sys\nsys.stdin=open("Input_1.txt","r")\nsys.stdout=open("Output_1.txt","w+")\n\n""")
        f1.write(code)
        f1.close()
        
        f2 = open("runner2.py","w+")
        f2.write("""import sys\nsys.stdin=open("Input_2.txt","r")\nsys.stdout=open("Output_2.txt","w+")\n\n""")
        f2.write(code)
        f2.close()
        
    def read_input1(self):
        input_1 = self.input1.get("1.0",END)
        f_inp1 = open("Input_1.txt","w+")
        f_inp1.write(input_1)
        f_inp1.close()
        
    def read_input2(self):
        input_2 = self.input2.get("1.0",END)
        f_inp2 = open("Input_2.txt","w+")
        f_inp2.write(input_2)
        f_inp2.close()
    
    def run(self):
        self.output1.delete("1.0",END)
        self.output2.delete("1.0",END)
        
        os.system("python runner1.py")
        os.system("python runner2.py")
        
        out1 = open("Output_1.txt","r")
        variable1 = out1.read()
        self.output1.insert(END,variable1)
        
        out2 = open("Output_2.txt","r")
        variable2 = out2.read()
        self.output2.insert(END,variable2)
        
IDE1 = PythonIDE()       
        
```
